```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
ex_image_file = '/content/ex_image_1.jpg'
image = cv2.imread(ex_image_file)
plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
image_rgb.shape
```

```python
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

```python
cropped_image_rgb = image_rgb[500:][200:800]
plt.imshow(cropped_image_rgb)
```

```python
cropped_image_gray = image_gray[500:][200:800]
plt.imshow(cropped_image_gray, cmap = 'gray')
```

```python
blur_image_rgb = cv2.GaussianBlur(image_rgb, (99,99), 0)
plt.imshow(blur_image_rgb)
```
